```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
file_2020 = "Resources/athletic_sales_2020.csv"
file_2021 = "Resources/athletic_sales_2021.csv"

as_2020_df = pd.read_csv(file_2020)
as_2021_df = pd.read_csv(file_2021)
```

```python
# Display the 2020 sales DataFrame
as_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
as_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
as_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
as_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
joined_df = pd.concat([as_2020_df, as_2021_df], ignore_index=True)

joined_df.head()
```

```python
# Check if any values are null.
joined_df.isnull().sum()
```

```python
# Check the data type of each column
joined_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
joined_df["invoice_date"] = pd.to_datetime(joined_df["invoice_date"])
```

```python
# Confirm that the "invoice_date" data type has been changed.
joined_df['invoice_date'].dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
region_df = joined_df.groupby(["region", "state", "city"])

products_sold = region_df["units_sold"].sum().reset_index(name="Total_Products_Sold")

products_sold.sort_values("Total_Products_Sold", ascending=False, inplace=True)

# Show the top 5 results.
products_sold.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_df = joined_df.pivot_table(index=["region", "state", "city"], values="units_sold", aggfunc="sum").reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_df = pivot_df.rename(columns={"units_sold": "Total_Products_Sold"})

pivot_df.sort_values("Total_Products_Sold", ascending=False, inplace=True)

# Show the top 5 results.
pivot_df.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales = region_df["total_sales"].sum().reset_index(name="Total Sales")

total_sales.sort_values("Total Sales", ascending=False, inplace=True)

# Show the top 5 results.
total_sales.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
sales_pivot_df = joined_df.pivot_table(index=["region", "state", "city"], values="total_sales", aggfunc="sum").reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
sales_pivot_df.rename(columns={"total_sales": "Total Sales"}, inplace=True)

sales_pivot_df.sort_values("Total Sales", ascending=False, inplace=True)

# Show the top 5 results.
sales_pivot_df.head(5)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
retailer_df = joined_df.groupby(["retailer", "region", "state", "city"])

retailer_sales = retailer_df["total_sales"].sum().reset_index(name="Total Sales")

retailer_sales.sort_values("Total Sales", ascending=False, inplace=True)

# Show the top 5 results.
retailer_sales.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
retailer_pivot_df = joined_df.pivot_table(index=["retailer", "region", "state", "city"], values="total_sales", aggfunc="sum").reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
retailer_pivot_df.rename(columns={"total_sales": "Total Sales"}, inplace=True)

retailer_pivot_df.sort_values("Total Sales", ascending=False, inplace=True)

# Show the top 5 results.
retailer_pivot_df.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_df = joined_df[joined_df["product"] == "Women's Athletic Footwear"]

womens_df.sort_values("total_sales", ascending=False, inplace=True)

# Show the top 5 results.
womens_df.head(5)
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_by_retailer = womens_df.groupby(["retailer", "region", "state", "city"])

womens_units_sold = womens_by_retailer["units_sold"].sum().reset_index(name="Womens_Footwear_Units_Sold")

womens_units_sold.sort_values("Womens_Footwear_Units_Sold", ascending=False, inplace=True)

# Show the top 5 results.
womens_units_sold.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_pivot_df = womens_df.pivot_table(index=["retailer", "region", "state", "city"], values="units_sold", aggfunc="sum").reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_pivot_df.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"}, inplace=True)

womens_pivot_df.sort_values("Womens_Footwear_Units_Sold", ascending=False, inplace=True)

# Show the top 5 results.
womens_pivot_df.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_day_sales = womens_df.pivot_table(index="invoice_date", values="total_sales", aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
womens_day_sales.rename(columns={"total_sales": "Total Sales"}, inplace=True)

womens_day_sales.sort_values("Total Sales", ascending=False, inplace=True)

# Show the table.
womens_day_sales.head(5)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = womens_day_sales.resample("D").sum()
## used github copiolt to assist in understanding the resample function

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_sales = daily_sales.sort_values(by="Total Sales", ascending=False)

daily_sales.sort_values("Total Sales", ascending=False, inplace=True)

# Show the top 5 results.
daily_sales.head(5)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = womens_day_sales.resample("W").sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
weekly_sales = weekly_sales.sort_values(by="Total Sales", ascending=False)

weekly_sales.sort_values("Total Sales", ascending=False, inplace=True)

# Show the top 5 results.
weekly_sales.head(5)
```
